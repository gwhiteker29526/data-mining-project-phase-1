```python
# ============================================
# Deliverable 1: Data Collection, Cleaning, and Exploration
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")

# Load dataset
df = pd.read_csv("data/diabetes_012_health_indicators_BRFSS2015.csv")

# Basic inspection
print("Shape:", df.shape)
display(df.head())
display(df.info())
display(df.describe())

# Check missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Check duplicates
duplicate_count = df.duplicated().sum()
print("\nDuplicate rows:", duplicate_count)

# Remove duplicates if needed
df = df.drop_duplicates()

print("\nShape after duplicate removal:", df.shape)

# -----------------------------
# Data quality checks
# -----------------------------
# Example: confirm expected ranges for selected binary columns
binary_cols = [
    'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke',
    'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
    'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk',
    'Sex'
]

for col in binary_cols:
    unique_vals = sorted(df[col].dropna().unique())
    print(f"{col}: {unique_vals}")

# Check target values
print("\nTarget values:", sorted(df['Diabetes_012'].unique()))

# -----------------------------
# Exploratory Data Analysis
# -----------------------------

# 1. Target distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Diabetes_012')
plt.title("Distribution of Diabetes Status")
plt.xlabel("Diabetes_012")
plt.ylabel("Count")
plt.show()

# 2. Numeric feature distributions
numeric_cols = ['BMI', 'MentHlth', 'PhysHlth', 'Age', 'Education', 'Income']

df[numeric_cols].hist(figsize=(12, 8), bins=25)
plt.suptitle("Distributions of Selected Numeric Features", y=1.02)
plt.tight_layout()
plt.show()

# 3. Boxplot for BMI by diabetes class
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Diabetes_012', y='BMI')
plt.title("BMI by Diabetes Status")
plt.show()

# 4. Boxplot for physical health by diabetes class
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Diabetes_012', y='PhysHlth')
plt.title("Physical Health Days by Diabetes Status")
plt.show()

# 5. Correlation heatmap
plt.figure(figsize=(14,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()

# 6. Compare means by diabetes status
grouped_means = df.groupby('Diabetes_012')[numeric_cols].mean()
display(grouped_means)

# 7. Relationship between age and BMI colored by diabetes class
sample_df = df.sample(5000, random_state=42)  # sample for faster plotting
plt.figure(figsize=(8,6))
sns.scatterplot(data=sample_df, x='Age', y='BMI', hue='Diabetes_012', alpha=0.6)
plt.title("Age vs BMI by Diabetes Status")
plt.show()
```

Dataset justification:
I selected the Diabetes Health Indicators dataset because it contains over 250,000 records and more than 20 attributes, satisfying the project requirements. The dataset includes demographic, behavioral, and health-condition features that are suitable for data cleaning, exploration, predictive modeling, clustering, and association analysis.

Cleaning discussion:
I checked for missing values, duplicate records, and unexpected values in binary columns. Duplicate rows were removed to reduce redundancy. I also validated that binary features only contained expected 0/1 values and confirmed the target classes in Diabetes_012.
